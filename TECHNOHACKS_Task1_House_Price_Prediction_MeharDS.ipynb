## Create X (predictor ) and y (target) variable:

```python
X = final_df.drop('price', axis=1)
y= final_df['price']
```

```python
X.head()
```

```python
y.head()
```

## Split the data into test and train:

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

## Model Building:


```python
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
rfr = RandomForestRegressor()
br = BaggingRegressor()
gbr = GradientBoostingRegressor()
adr = AdaBoostRegressor()
lr = LinearRegression()
knr = KNeighborsRegressor()
svr= SVR()
dtg = DecisionTreeRegressor()
```

```python
clfs={
    'Random Forest Regressor':rfr,
    'Bagging Regressor': br,
    'Gradient Boosting Regressor':gbr,
    'Ada Boost Regressor': adr,
    'Linear Regression':lr,
    'K Neighbors Regressor':knr,
    'SVR':svr,
    'Decision Tree Regressor':dtg
}
```

## Model Evaluation:
Calculate :> Score, mean_absolute_error, mean_squared_error, r2_score

```python
# Function for Calculate Score, mean_absolute_error, mean_squared_error, r2_score
def train_classifer(clf,X_train, X_test, y_train, y_test):
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    Score = clf.score(X_test,y_test)
    MeanAbsoluteError= mean_absolute_error(y_test, y_pred)
    MeanSquaredError = mean_squared_error(y_test, y_pred)
    R2score = r2_score(y_test, y_pred)
    
    return Score, MeanAbsoluteError, MeanSquaredError, R2score
```

```python
score=[]
MeanAbsoluteError =[]
MeanSquaredError = []
R2score= []
for name, clf in clfs.items():
    current_Score, current_Absolute_error, current_Squared_error, current_r2score = train_classifer(clf,X_train, X_test, y_train, y_test)
    print('The performance evaluation of {} :'.format(name))
    print('Score :', current_Score)
    print('Mean Absolute Error :',current_Absolute_error)
    print('Mean Squared Error :', current_Squared_error)
    print('R2 score :',current_r2score )
    print('\n' , '-'*50, '\n')
    score.append(current_Score)
    MeanAbsoluteError.append(current_Absolute_error)
    MeanSquaredError.append(current_Squared_error)
    R2score.append(current_r2score)
```

## Creating Performance Data Frame:


```python
performace_df = pd.DataFrame({'Algorithm': clfs.keys(), 'Score': score, 'Mean Absolute Error': MeanAbsoluteError, 'Mean Squared Error':MeanSquaredError, 'R2 score':R2score})
```

```python
performace_df
```

## pandas.melt:
This function is useful to massage a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns, considered measured variables (value_vars), are “unpivoted” to the row axis, leaving just two non-identifier columns, ‘variable’ and ‘value’.

```python
performace_df1=pd.melt(performace_df, id_vars='Algorithm')
```

```python
performace_df1
```

## Lets Plot Evaluation Matrix:


```python
sns.catplot(x='Algorithm', y='value', hue='variable', data=performace_df1, kind='bar', height=5)
plt.ylim(0.5,1.0)
plt.xticks(rotation='vertical')
plt.show()
```
